```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://www.worldometers.info/geography/how-many-countries-are-there-in-the-world/#google_vignette"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html')
print(soup)
```

```python
table_element = soup.find('table', class_='row')
```

```python
table = soup.find('table')
Headings = table.find_all('th')
```

```python
print(Headings)
```

```python
Header = [i.text for i in Headings]
print(Header)
```

```python
df = pd.DataFrame(columns = Header)
df
```

```python
rows = table.find_all('tr')
print(rows)
print(df.columns)
```

```python
ind=1
for i in rows[1:11]:
  row_data = i.find_all('td')
  every_row_data = [i.text for i in row_data]
  print(every_row_data)
  df.loc[ind] = every_row_data
  ind = ind+1
```

```python
df
df.to_csv("Allcountries.csv")
```
